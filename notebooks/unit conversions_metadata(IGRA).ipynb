```python
import pandas as pd
import numpy as np
import csv
```

```python
metadata_IGRA= pd.read_csv('/Users/koushal/Downloads/weather_balloon_metadata.csv')
metadata_IGRA
```

```python
def celsius_to_kelvin(celsius):
    kelvin = celsius + 273.15
    return kelvin

columns_to_convert = ['LI','SI','KI','TTI']
for col in columns_to_convert:
    new_col_name = col + '(F)'
    metadata_IGRA[new_col_name] = metadata_IGRA[col].apply(celsius_to_kelvin)
metadata_IGRA = metadata_IGRA.drop(columns=columns_to_convert)
```

```python
def meters_to_feet(meters):
    feet = meters * 3.28084
    return feet
metadata_IGRA['INVHGT'+'(Ft)'] = metadata_IGRA['INVHGT'].apply(meters_to_feet)
metadata_IGRA['LFCHGT'+'(Ft)']= metadata_IGRA['LFCHGT'].apply(meters_to_feet)
metadata_IGRA['MIXHGT'+'(Ft)']= metadata_IGRA['MIXHGT'].apply(meters_to_feet)
metadata_IGRA['FRZHGT'+'(Ft)']= metadata_IGRA['FRZHGT'].apply(meters_to_feet)
metadata_IGRA['LCLHGT'+'(Ft)']= metadata_IGRA['LCLHGT'].apply(meters_to_feet)
metadata_IGRA['LNBHGT'+'(Ft)']= metadata_IGRA['LNBHGT'].apply(meters_to_feet)
metadata_IGRA=metadata_IGRA.drop(['INVHGT','LFCHGT','MIXHGT','FRZHGT','LCLHGT','LNBHGT'],axis=1)
```

```python
def kelvin_to_fahrenheit(kelvin):
    fahrenheit = (kelvin/10 - 273.15) * 9/5 + 32
    return fahrenheit
metadata_IGRA['INVTEMPDIF'+'(F)'] = metadata_IGRA['INVTEMPDIF'].apply(kelvin_to_fahrenheit)
```
